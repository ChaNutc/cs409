```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-1,1,0.001)
yt=3*x+2+(np.random.randn(len(x)))
y=3*x+2
plt.plot(x,yt,'o')
plt.plot(x,y,'r')
```

$
L=\frac{1}{2n}\sum_i^n(y_p^i-y_t^i)^2
$


$
\theta_1 = \theta_1 - \alpha \frac{\partial L}{\partial\theta}$

```python
theta1 = 0.0
theta2 = 0.0
alpha = 0.1

def yp(x):
    global theta1,theta2
    return theta1*x+theta2
plt.plot(x,yt,'o')
for i in range(100):
    theta1 = theta1 - alpha*np.average((yp(x) - yt)*x)
    theta2 = theta2 - alpha*np.average((yp(x) - yt))
    if i%10==0 :
        print(theta1,theta2)
        plt.plot(x,theta1*x+theta2,'r')
```

```python
n=200
def normal(n):
    return np.random.randn(n)
x1=np.concatenate((normal(n)-2,normal(n)+2),axis=None)
x2=np.concatenate((normal(n)-2,normal(n)+2),axis=None)
yt=np.ones(2*n)
yt[:n]=-1
plt.plot(x1[yt==-1],x2[yt==-1],'or')
plt.plot(x1[yt==1],x2[yt==1],'ob')
```

```python
theta1 = 0.0
theta2 = 0.0
theta3 = 0.0
alpha = 0.1
step = 0.1
mx2,mx1 = np.mgrid[slice(-5,5+step,step),
                    slice(-5,5+step,step)]
def func_yp(x1,x2):
    global theta1,theta2,theta3
    z=theta1*x1 + theta2*x2 + theta3
    return np.tanh(z)
plt.plot(x1[yt==-1],x2[yt==-1],'or')
plt.plot(x1[yt==1],x2[yt==1],'ob')
for i in range(100):
    yp = func_yp(x1,x2)
    theta1 = theta1 - alpha*np.average((yp - yt)*(1-pow(yp,2))*x1)
    theta2 = theta2 - alpha*np.average((yp - yt)*(1-pow(yp,2))*x2)
    theta3 = theta3 - alpha*np.average((yp - yt)*(1-pow(yp,2)))
    if i%10==0 :
        print(theta1,theta2,theta3)
        #plt.plot(x,theta1*x1+theta2*x2+theta3,'r')
cmap = plt.get_cmap('RdYlBu')
plt.pcolormesh(mx1,mx2,func_yp(mx1,mx2),cmap=cmap,vmin=-2,vmax=1.5)
```
